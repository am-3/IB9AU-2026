## Objective
This project aims to conduct a comparative "Risk Divergence" analysis between Block, Inc. and PayPal Holdings, Inc. based on their respective 2023 10-K "Item 1A. Risk Factors" disclosures. The core question investigated is whether Block's strategic pivot towards "ecosystem expansion" (e.g., Bitcoin, Web3, AI) is generating disproportionately higher, potentially undisclosed operational and regulatory risks compared to PayPal's more conservative "stable incumbent" approach. The analysis employs advanced prompting techniques with a large language model (LLM) to identify subtle risk differentiators and concludes with a synthesized verdict for investors.
___
## Tech Stack
*   **Programming Language**: Python 3.x
*   **Libraries**: `PyPDF2` for PDF text extraction, `re` for regular expressions, `google.generativeai` for LLM-powered analysis.
*   **LLM**: Google Gemini (specifically `gemini-2.5-flash`).
*   **Environment**: Google Colab.
___
## Methodology
The methodology involves a multi-pronged approach to risk assessment:
1.  **Data Ingestion**: Extracting "Item 1A. Risk Factors" from 2023 10-K PDF documents for both companies.
2.  **LLM-driven Analysis**: Applying various prompting techniques (Zero-Shot, Persona-based Regulatory Audit, Chain-of-Thought, Dialectical Debate) to the extracted risk factors using the Gemini LLM.
3.  **LLM-as-a-Judge Evaluation**: A "Director of Quantitative Risk" persona (LLM) critically evaluates the output from each prompting technique based on Evidence Quality, Regulatory Nuance, and Actionability for Investors.
4.  **Synthesized Executive Summary**: Based on the LLM-as-a-Judge's feedback, a final executive summary is generated to provide a definitive answer to the base question, highlighting key regulatory and operational risk domains.

```python
import google.generativeai as genai
from google.colab import userdata
import PyPDF2
import re

# Used to securely store your API key
# Ensure 'GEMINI_API_KEY' is set in Colab secrets for secure access.
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# Initialize the Gemini API with a specified model.
gemini_model = genai.GenerativeModel('gemini-2.5-flash')

print("Gemini API configured successfully.")

def extract_text_from_pdf(pdf_path):
    """
    Extracts text content from a PDF file.

    Args:
        pdf_path (str): The file path to the PDF document.

    Returns:
        str: The extracted text from the PDF, or an error message if extraction fails.
    """
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ''
            for page in reader.pages:
                # Append extracted text; handle None if page.extract_text() returns nothing.
                text += page.extract_text() if page.extract_text() else ''
            return text
    except Exception as e:
        return f"Error reading PDF {pdf_path}: {e}"

def extract_item_1a_risk_factors(full_text):
    """
    Extracts the 'Item 1A. Risk Factors' section from the full text of a 10-K report.

    The extraction is performed by identifying "Item 1A. Risk Factors" and
    capturing all subsequent text until the next "Item X." or end of document.

    Args:
        full_text (str): The entire text content of the 10-K report.

    Returns:
        str: The extracted 'Item 1A. Risk Factors' section, or a message indicating
             that the section could not be found.
    """
    # Regex to find "Item 1A. Risk Factors" and the subsequent section
    # This regex looks for "Item 1A. Risk Factors" and captures until the next "Item X." or end of document
    # Uses re.DOTALL to match across newlines and re.IGNORECASE for case-insensitive matching.
    item_1a_regex = r'Item\s*1A\.\s*Risk\s*Factors(.*?)(?=Item\s*[1-9]\.[^\d]|Item\s*[A-Z]\.|\Z)'
    match = re.search(item_1a_regex, full_text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(1).strip()
    return "Could not find 'Item 1A. Risk Factors' section or subsequent item." # Clearer error message

# Define PDF paths - ensure these files are uploaded to the Colab environment
PAYPAL_10K_PATH = '/content/PayPal 10K 2023.pdf'
BLOCK_10K_PATH = '/content/Block 10K 2023.pdf'

# Load full text from PDFs
paypal_10k_full_text = extract_text_from_pdf(PAYPAL_10K_PATH)
block_10k_full_text = extract_text_from_pdf(BLOCK_10K_PATH)

# Extract 'Item 1A. Risk Factors' section
paypal_10k_risk_factors = extract_item_1a_risk_factors(paypal_10k_full_text)
block_10k_risk_factors = extract_item_1a_risk_factors(block_10k_full_text)

print(f"PayPal 10-K Risk Factors length: {len(paypal_10k_risk_factors)} characters")
print(f"Block 10-K Risk Factors length: {len(block_10k_risk_factors)} characters")

# Add warnings if risk factors were not successfully extracted.
if "Could not find 'Item 1A. Risk Factors'" in paypal_10k_risk_factors:
    print("\nWarning: PayPal 10-K 'Item 1A. Risk Factors' not extracted. Please ensure the PDF is correctly uploaded and the extraction regex is suitable for its format.")
if "Could not find 'Item 1A. Risk Factors'" in block_10k_risk_factors:
    print("\nWarning: Block 10-K 'Item 1A. Risk Factors' not extracted. Please ensure the PDF is correctly uploaded and the extraction regex is suitable for its format.")
```

# Insights & Technical Learnings

## Key Results
*   **Data Extraction Challenge**: A significant hurdle was the initial failure to accurately extract the full "Item 1A. Risk Factors" text from the 10-K PDFs using the provided regex. This highlights the fragility of regex-based parsing for varied document structures and the critical need for robust data ingestion.
*   **LLM's Qualitative Strength**: Despite the missing detailed textual evidence, the "Persona / Regulatory Auditor" (Response 2) prompting technique excelled in providing deep, nuanced qualitative insights. It demonstrated the LLM's capacity to infer and structure expert analysis based on its training data and the known business models, effectively outlining **expected** risk areas for Block's crypto and AI ventures vs. PayPal's traditional operations.
*   **Quantitative Signal of Regulatory Scrutiny**: The "Debate / Dialectical" (Response 4) uniquely leveraged quantitative data—specifically, the page disparity in "Item 1B. Unresolved Staff Comments" (Block: 61 pages vs. PayPal: 31 pages)—to identify a subtle yet alarming risk divergence. This numerical difference was interpreted as direct evidence of higher, unresolved regulatory scrutiny for Block, an actionable signal for investors.
___
## Technical Learnings
*   **Robust Data Ingestion is Paramount**: This task underscored that the quality of LLM analysis is fundamentally bottlenecked by the quality and completeness of its input data. Future iterations would benefit from more sophisticated PDF parsing libraries (e.g., `camelot`, `pdfplumber`) or direct SEC Edgar API integration for reliable, structured data extraction.
*   **Strategic Prompting for LLM Capabilities**: Different prompting techniques unlocked distinct analytical strengths of the LLM:
    *   **Persona-based Prompting**: Effective for generating expert qualitative assessments and identifying *conceptual* risk divergences.
    *   **Dialectical Prompting**: Proved adept at leveraging subtle quantitative cues (like document section lengths) to infer deeper implications and identify 'unpriced' risks, especially when direct content was limited.
*   **LLM as a 'Sense-Maker' vs. 'Fact-Retriever'**: The exercise clarified that, without explicit textual evidence, LLMs can plausibly generate content. The LLM-as-a-Judge mechanism was crucial for penalizing unsubstantiated claims and emphasizing evidence-based reasoning, reinforcing the need for guardrails against hallucination.
